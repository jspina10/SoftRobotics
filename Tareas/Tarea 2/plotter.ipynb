```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('particle_1_x.txt', delim_whitespace=True, header=None, comment='#')

# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.plot(data.iloc[:, 0], data.iloc[:, 2], label='Y coordinate')
plt.plot(data.iloc[:, 0], data.iloc[:, 1], label='X coordinate')
plt.plot(data.iloc[:, 0], data.iloc[:, 3], label='Z coordinate')
plt.title('Posición de la pelota en el tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (mm)')
plt.legend()
plt.grid(True)
plt.show()
```
