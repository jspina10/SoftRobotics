```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data, skipping the initial comments and handling spaces as separators
data = pd.read_csv('particle_1_f.txt', delim_whitespace=True, header=None, comment='#', 
                   names=['time', 'placeholder', 'y', 'z'])

# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
plt.plot(data['time'], data['y'], label='Y coordinate')
plt.title('Y Coordinate over Time')
plt.xlabel('Time')
plt.ylabel('Y Coordinate')
plt.legend()
plt.grid(True)
plt.show()
```
